```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.datasets import mnist
import numpy as np
from tensorflow.keras.utils import to_categorical
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
print("Shape of X_train is ",X_train.shape)
print("Shape of X_test is ",X_test.shape)
```

```python
#Reshaping

X_train.reshape(-1,28,28,1)
X_test.reshape(-1,28,28,1)
```

```python
#Normalization

X_train = X_train.astype('float32')/255
X_test = X_test.astype('float32')/255
```

```python
#Label Encoding

y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
```

```python
#Building the CNN Model

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

model = Sequential()
model.add(Conv2D(32, (3,3), activation = 'relu', input_shape = (28,28,1)))
model.add(MaxPooling2D((2,2)))

model.add(Flatten())

model.add(Dense(64, activation = 'relu'))
model.add(Dense(10, activation = 'softmax'))


model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])












```

```python
#Training The Model


history  = model.fit(X_train, y_train, epochs = 10, batch_size = 32, validation_split = 0.2, verbose = 1)
```

```python
# 1. Test data par accuracy check karna
test_loss, test_acc = model.evaluate(X_test, y_test, verbose=0)
print(f"Test Accuracy: {test_acc*100:.2f}%")

# 2. Prediction check karna (Pehli image par)
import numpy as np
sample_image = X_test[0].reshape(1, 28, 28, 1) # Model ko 4D input chahiye
prediction = model.predict(sample_image)
predicted_label = np.argmax(prediction) # Sab se bari probability wala digit

print(f"Model says this is a: {predicted_label}")
print(f"Actual digit was: {np.argmax(y_test[0])}")
```
